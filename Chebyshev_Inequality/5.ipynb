```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

sns.set(style="ticks", context="talk")
plt.style.use("dark_background")


from scipy.stats import binom, poisson, norm
```

# Binomial and Normal Distributions

```python
n, p = 16, 0.5
X = 10
```

### Calculating Probability of Binomial Distribution

```python
print("Probability of 10 Heads in 16 Coin Flips : " + str(binom.pmf(k = 10, p = 0.5, n = 16, loc = 0)))
```

### Approximating Probability using Normal Distribitution

```python
print('n * p = ' + str(n*p) + ' > 5'
      + '\nn * (1-p) = ' + str(n * (1-p)) + ' > 5\n\n')
print('Both Values are more than 5, Therefore We Can use Normal Approximation\n\n')

mean = n * p
std = (n * p * (1 - p)) ** 0.5
print('mean = ' + str(mean) + '\nstandard deviation = ' + str(std) + '\n\n')

print('P(X = 10) is equivalent to P(9.5 < X < 10.5)')

z_score_1 = (X + 0.5 - mean) / std
z_score_2 = (X - 0.5 - mean) / std
print('z_score_1 = ' + str(z_score_1) + '\nz_score_2 = ' + str(z_score_2) + '\n\n')


rv = norm(0, 1)
p_norm = abs(rv.cdf(z_score_2) - rv.cdf(z_score_1))

print('Probability of ' + str(z_score_2) + ' < X < ' + str(z_score_1) + ' in a normal distributed data is : ' +
     str(p_norm))
```

### Comparing the Results

```python
print("Probability of 10 Heads in 16 Coin Flips : " + str(binom.pmf(k = 10, p = 0.5, n = 16, loc = 0)))
print('Probability of ' + str(z_score_2) + ' < X < ' + str(z_score_1) + ' in a normal distributed data is : ' +
     str(p_norm))
```

# Plot

```python
fig1, ax1 = plt.subplots(figsize = (12, 8))

data_binom = binom.rvs(n = n, p = p, loc = 0, size = 200)
sns.distplot(data_binom, kde = True)

data_normal = np.random.normal(data_binom.mean(), data_binom.std(), size = 200)
sns.distplot(data_normal, kde = True)

ax1.legend(labels = ['Binomial Distribution', 'Normal Distribution'])
```
