```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="ticks", context="talk")
plt.style.use("dark_background")

from scipy.stats import binom, poisson, norm
```

```python
k = np.arange(1, 5, 1)
```

```python
def chebyshev_lbound(k):
    return (1 - (1 / k ** 2))
```

# Binomial Distribution

```python
n, p = 40, 0.8
mean = binom.mean(n = n, p = p, loc = 0)
std = binom.std(n = n, p = p, loc = 0)
```

```python
def delta_cdf_binom(k):
    return binom.cdf(k = k * std + mean, n = n, p = p) - binom.cdf(k = -k * std + mean, n = n, p = p)
```

```python
fig1, ax1 = plt.subplots(figsize = (12, 8))

plt.plot(k, chebyshev_lbound(k))
plt.plot(k, delta_cdf_binom(k))
ax1.legend(title = 'Binomial Distribution with n = 40, p = 0.8' , labels=["Cherishef ","CDF"])
```

# Poisson Distribution

```python
mu = 0.6
mean = poisson.mean(mu = mu, loc = 0)
std = poisson.std(mu = mu, loc = 0)
```

```python
def delta_cdf_poisson(k):
    return poisson.cdf(k = k * std + mean, mu = mu, loc = 0) - poisson.cdf(k = -k * std + mean, mu = mu, loc = 0)
```

```python
fig2, ax2 = plt.subplots(figsize = (12, 8))

plt.plot(k, chebyshev_lbound(k))
plt.plot(k, delta_cdf_poisson(k))
ax2.legend(title = 'Poisson Distribution with mu = 0.6', labels=["Cherishef ","CDF"])
```

# Normal Distribution

```python
mean = norm.mean(loc = 0, scale = 1)
std = norm.std(loc = 0, scale = 1)
```

```python
def delta_cdf_norm(k):
        return norm.cdf(x = k * std + mean, loc = 0, scale = 1) - norm.cdf(x = -k * std + mean, loc = 0, scale = 1)
```

```python
fig3, ax3 = plt.subplots(figsize = (12, 8))

plt.plot(k, chebyshev_lbound(k))
plt.plot(k, delta_cdf_norm(k))
ax3.legend(title = 'Normal Distribution with mean = 0, scale = 1', labels=["Cherishef ","CDF"])
```
